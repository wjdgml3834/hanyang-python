**< 복습문제2: 글자수 마을 >**

글자수 마을은 음식의 가격이 글자수로 정해지는 마을입니다. 글자수 마을에는 여러 요리사들이 있는데 각자 만들 수 있는 음식이 다릅니다. 요리사의 이름과 만들 수 있는 요리 리스트가 주어질 때, 가장 비싼 음식을 만드는 요리사의 이름과 요리명을 출력하는 프로그램을 제작하시오.

< 힌트 >
- 반복문을 이용해 인덱스 접근 range(len())
- 요리의 이름 길이를 새로운 리스트에 저장 (list()를 이용해 생성 후, append 이용)
- .index()를 이용하여 리스트에서 특정 값의 인덱스를 찾기
- 리스트의 최대값은 max()를 이용해 구할 수 있습니다.

```python
name = ['나현', '수현', '제인', '봉준', '혜민', '정희', '현우', '종민', '민기', '주연', '담비', '가은']
cook = ['오징어짬뽕', '너구리', '튀김소보루', '연어김밥', '족발튀김', '감바스', '라멘', '새우찜', '조개탕', '까르보나라', '불고기덮밥', '비빔밥']

# TODO

cook_length = list()

for i in range(len(name)):
    cook_length.append(len(cook[i]))

for i in range(len(cook_length)):
    if(cook_length[i] == max(cook_length)):
        print(f'요리사: {name[i]} , 요리명: {cook[i]}')





```

**2-1. 홀수인지 검사하는 프로그램을 만들어보자**

```python
# 함수 선언을 한 아래에 실행 문장 작성
# 매개변수를 사용한 연산
# 입력받는 매개변수의 개수를 고려

def is_odd(a): # 홀수인지 검사하는 함수를 만들어보자
  if a % 2 == 1: # 2로 나눴을 때 나머지가 1이라면 = 홀수는 2로 나누면 1이 남으니까
    return f"{a}는 홀수입니다." # 이 문장을 함수의 결과값으로 반환해라
  else: # 2로 나눴을 때 나머지가 1이 아니라면 = 2로 나눴을 때 나머지는 0 아니면 1. 즉 1이 아니라면 0일 수밖에 없음. = 짝수
    return f"{a}는 짝수입니다."
```

```python
# 선언한 함수를 실행할 때는 인자를 넣어서

is_odd(8)
```

**문자열 대/소문자 변환 : upper, lower, title, capitalize**

upper은 모든 문자를 대문자로,

lower은 모든 문자를 소문자로,

```python
text = 'please BE my eternity, please call my name. Run away, run away, run away.'

print(text.upper())

print(text.lower())

print(text.capitalize())
```

```python
best_food = {'korea':'jokbal',
             'japan':'sushi',
             'china':'Jajangmyeon'}

print(best_food["korea"])
print(best_food.keys())
print(best_food.values())
print(best_food.items())

# 요렇게도 출력 가능 
for key,value in best_food.items():
    print(key,value)
```

```python
name = ['나현', '수현', '제인', '봉준', '혜민', '정희', '현우', '종민', '민기', '주연', '담비', '가은']
cook = ['오징어짬뽕', '너구리', '튀김소보루', '연어김밥', '족발튀김', '감바스', '라멘', '새우찜', '조개탕', '까르보나라', '불고기덮밥', '비빔밥']

dict_list = dict() # set이랑 헷갈리니까 {} 보다는 요렇게 선언해주기.
for i in range(len(name)):
    dict_list[name[i]] = cook[i]
    
print(dict_list)
```
