## **1. 조 정하기**
- 학생 이름이 있을 때, 두명씩 짝지어서 랜덤하게 조를 정하려 합니다.
- 랜덤하게 두명씩 짝을 지어 출력하는 프로그램을 작성하시오.
- random.shuffle(리스트) 를 이용하면 리스트를 랜덤하게 섞을 수 있습니다.
- 리스트를 한번 shuffle한 뒤, 앞에서부터 두명씩 출력한다면, 랜덤하게 두명씩 짝을 지을 수 있습니다.

```python
# 일단 팀부터 짜볼까요?
import random
random.seed(1027) 
student = '김가은, 김수현, 김정희, 김주연, 김현우, 김혜민, 문봉준, 박민기, 변제인, 이나현, 이담비, 이종민'

student_list = student.split(', ')

random.shuffle(student_list)

print(student_list)


for i in range(0,len(student_list) - 1,2):
    print(student_list[i], student_list[i+1])



```

## **6. 눈치껏 규칙 찾아보기**
- https://humanities.hanyang.ac.kr/front/community/notice
- 1-5페이지를 추출하려면 어떻게 해야할까요?

```python
# TODO

# 주소 url에 페이지가 1,2 이렇게 바뀐다.

for page in range(1,6):
    url ='https://humanities.hanyang.ac.kr/front/community/notice'.format(page)
    res= requests.get(url)
    soup = BeautifulSoup(res.text, 'html.parser')
    for t in soup.find_all('span', class_='subject'):
        output = t.text
        output = output.split('								')[1] # 공백은 직접 출력해보고 공백 복사해주기 
        output= output.split('				')[0]
        print(output)

```

## **7.위치로 추출하기**
- https://humanities.hanyang.ac.kr/front/introduce/status
- 교수님들의 이름만 추출해 봅시다.

```python
# 선생님 풀이 

import requests

response = requests.get('https://humanities.hanyang.ac.kr/front/introduce/status')

from bs4 import BeautifulSoup 

soup = BeautifulSoup(response.text, 'html.parser')

for tr in soup.find_all('tr'):
    if str(tr).find('<th>') >= 0: continue # th만 건너 띄워라. 
    td_list = tr.find_all('td')
    print(td_list[1].text)
```
