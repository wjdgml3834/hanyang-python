## **2. GET과 POST**
- 서버에 정보를 전달할 때는 두 가지 방법을 사용할 수 있습니다: GET과 POST
- 일반적으로 get은 정보를 가져올 때, post는 서버 내에 정보를 바꿀 때 사용합니다.
- GET은 URL에 정보를 넣어서 전달하는 방법입니다. 사용이 간단하지만, 길이가 긴 내용은 전달이 어렵습니다. 보통 ? 뒤에 &로 구분하여 정보를 넣습니다.
 - http://abcdefgabcdefg/abcdefg.php?a=에이&b=비&c=3
 - abcdefg.php라는 곳으로, a는 에이, b는 비, c는 3이라는 정보가 전달됩니다.
- POST는 전송해야될 데이터를 HTTP 메세지의 Body에 담아서 전송합니다.


**1) 데이터 전달 예제(GET/POST)**

```python
import requests

url = 'https://httpbin.org/get'

http_post_request = requests.get(url, params = {"name": "james"}) # dict 형태이다. &기준으로 나뉜다.

print(http_post_request.text)
print(http_post_request.status_code)
```

```python
import requests

url = 'https://httpbin.org/get&?name=james' # url에 직접전달도 가능

http_post_request = requests.get(url)

print(http_post_request.text)
print(http_post_request.status_code)
```

```python
import requests

url = 'https://httpbin.org/post'
data = {'name': 'james',
        'age': '23',
        }

http_post_request = requests.post(url, json = data) # json이라는 인수를 주고 data.

print(http_post_request.text)
print(http_post_request.status_code)
```

**2-1) GET을 이용해서 자신의 이름과 나이를 httpbin으로 전달해봅시다.**

```python
import requests

url = 'https://httpbin.org/get'

http_post_request = requests.get(url, params = {"name": "jung hee", "age": "27"}) # dict 형태이다. &기준으로 나뉜다.

print(http_post_request.text)
print(http_post_request.status_code)
```

**2-2) POST을 이용해서 자신의 이름과 나이를 httpbin으로 전달해봅시다.**

```python
import requests

url = 'https://httpbin.org/post'
data = {'name': 'jung hee',
        'age': '27',
        }

http_post_request = requests.post(url, json = data) # json이라는 인수를 주고 data.

print(http_post_request.text) # origin: 어떤 컴퓨터에서 나에게 정보를 보냈느냐? useragent: python requset를 바꾸는 경우가 많다.
# 헤더에는 접속한 브라우저의 정보가 담겨있다. 
print(http_post_request.status_code)
```

## **3. Header**
- 내가 넣은 정보는 body에..
- header란 무엇일까요?

HTTPRequestMessageFormat.png

```python
import requests

url = 'https://httpbin.org/post'
headers = {'User-Agent': 'Mozilla/5.0'}  # userAgent 바꾸기. -> 웬만하면 다 뚫린다. 정상적인 방법은 network탭에서 headers 탭을 들어가본다. 
# 그중에서 request headers를 본다. 
data = {'name': 'james',
        'age': '23',
        }

http_post_request = requests.post(url, headers=headers, json = data)

print(http_post_request.text)
print(http_post_request.status_code)
```

**과제2: 목요웹툰의 제목과 설명 수집하기**

https://comic.naver.com/webtoon/weekdayList?week=thu

- 네이버 목요웹툰에서 제목과 설명을 수집해서 dict로 저장합니다.
- {'연애혁명': '평범하면서 금사빠인 고등학생 순정남 공주영은 까칠하고 차가운 학생 왕자림을 보고 사랑에 빠져버린다. 너무 다른 둘, 괜찮을까?'}

```python
import requests
from bs4 import BeautifulSoup

res = requests.get('https://comic.naver.com/webtoon/weekdayList?week=thu')
soup = BeautifulSoup(res.text, 'html.parser')
detail_dict = dict()

for dt in soup.find_all('dt'):
    title_tag = dt.find_all('a')
    title = title_tag.text.strip()
    for title in title_tag:
        attrs = title.attrs['href']
        title_id = attrs[22:28]
        res_detail = requests.get(f'https://comic.naver.com/webtoon/list?titleId={title_id}&weekday=thu')
        soup = BeautifulSoup(res_detail.text, 'html.parser')
        for detail_div in soup.find_all('div', class_='detail'):
            detail_p = detail_div.find('p')
            detail_dict[title] = detail_p.text.strip()
            
print(detail_dict)
            

    
    
```
