## **10. 문자열 indexing & slicing**
- 문자열에 속한 문자들은 `순서`가 있습니다.
- 문자열의 `순서`를 `인덱스`라고 합니다.
- 인덱스는 0부터 시작합니다 (0번째 문자, 1번째 문자, 2번쨰 문자...)
- -1번째 인덱스는 제일 마지막 문자를 말합니다.
- [시작:끝]와 같이 명시하여 [시작, 끝)에 해당하는 부분 문자열을 추출
- 인덱스에 접근을 위해서는 []를 이용하면 됩니다.
 - a라는 문자열의 5번째 문자열: a[4]
 - b라는 문자열의 마지막 문자열: a[-1]
 - a라는 문자열의 3,4,5번째 문자열 슬라이싱: a[3:6]

그럼, 다음 문자열에서 인덱스 접근을 통해 바나나를 출력 해 볼까요?

```python
hanguls = "가나다라마바사아자차카타파하"

print(hanguls[0], hanguls[1], hanguls[2], hanguls[-1])
```

이번에는, 문자열 슬라이싱을 통해 ello를 출력해 봅시다.

```python
a = 'Hello, world!'

print(a[0:13])
```

## **12. 조건문(Condition)**
- 특정 조건을 만족하는 경우에만 작업을 수행
- 종류: if, elif, else
- if, elif, else에 종속된 코드는 들여쓰기로 표현(블록이라 부름)
- 파이썬에서 모든 블록의 시작줄은 :로 끝나야 함.

if 조건1:

　조건1이 맞을 때 실행 할 내용

elif 조건2:

　조건1은 만족하지 못하지만, 조건2는 만족할 때 실행 할 내용

else:

　모든 조건을 만족하지 못할 때 실행 할 내용
